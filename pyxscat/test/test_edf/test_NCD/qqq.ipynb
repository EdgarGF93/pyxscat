```python
from pyFAI import load
import matplotlib.pyplot as plt
from pyFAI.detectors import Pilatus1M
from pyFAI.azimuthalIntegrator import AzimuthalIntegrator
import numpy as np
```

```python
p1m = Pilatus1M()
```

```python
ai = AzimuthalIntegrator(detector=p1m, poni1=0.05, poni2=0.05)
```

```python
arr_tth = ai.array_from_unit(unit='2th_deg')
```

```python
plt.imshow(arr_tth)
plt.colorbar()
```

```python
np.max(arr_tth)
```

```python
ai2 = AzimuthalIntegrator(detector=p1m, poni1=0.05, poni2=0.1)
arr_tth_2 = ai2.array_from_unit(unit='2th_deg')
plt.imshow(arr_tth_2)
plt.colorbar()
plt.show()
```

```python
np.max(arr_tth_2)
```

```python
ai = load('Cr2O3/Cr2O3.poni')
```

```python
tth_arr = ai.array_from_unit(unit='2th_deg')
chi_arr = ai.array_from_unit(unit='chi_rad')
qx_arr = ai.array_from_unit(unit='qx_nm^-1')
qy_arr = ai.array_from_unit(unit='qy_nm^-1')
q_arr = ai.array_from_unit(unit='q_nm^-1')
```

```python
fig, axes = plt.subplots(ncols=3)
arrs = [qx_arr, qy_arr, chi_arr]
labels = ['Qx', 'Qy', 'chi']
for ind in range(3):
    axes[ind].imshow(arrs[ind])
    axes[ind].set_title(labels[ind])
```
